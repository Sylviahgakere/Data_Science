```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Load the data
Sylvia = pd.read_csv('/content/kenyan_accidents.csv')
# Extract the independent variable (X) and dependent variable (y)
X = Sylvia[['AGE']]
y = Sylvia['NO OF VICTIMS']
# Create and fit a linear regression model
regressor = LinearRegression()
regressor.fit(X, y)
# Plot the data points
Sylvia.plot(x='AGE', y='NO OF VICTIMS', style='o', label='Data Points')
# Overlay the regression line
plt.plot(X, regressor.predict(X), color='red', linewidth=2, label='Regression Line')
plt.xlabel('AGE')
plt.ylabel('NO OF VICTIMS')
plt.legend()
plt.show()
```
